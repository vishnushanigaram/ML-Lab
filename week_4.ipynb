```python
#9 Scaling the data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[["area", "bedrooms", "bathrooms"]] = scaler.fit_transform(df[["area", "bedrooms", "bathrooms"]])
print(df.head())

```

```python
from sklearn.model_selection import train_test_split

x=df.drop("price",axis=1)
y=df["price"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

```

```python
##11
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)

print("Coefficients: ", model.coef_)

print("Intercept: ", model.intercept_)

```

```python
##12
y_train_pred = model.predict(x_train)
errors_train = y_train_pred - y_train
```

```python
#13
plt.scatter(y_train_pred, errors_train)
plt.xlabel("Predicted Prices")
plt.ylabel("Errors")
plt.title("Residual Plot - Train Data")
plt.show()
```

```python
#14
y_test_pred = model.predict(x_test)
errors_test = y_test_pred - y_test
```

```python
#15
plt.scatter(y_test_pred, errors_test)
plt.xlabel("Predicted Prices")
plt.ylabel("Errors")
plt.title("Residual Plot - Test Data")
plt.show()
```

```python
#16
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_test_pred)
r2 = r2_score(y_test, y_test_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-Squared Value: {r2}")
```
